```python
import pandas as pd
import numpy as np
import glob
```

### Carregando as bases de dados manipuladas n1

```python
df_carga : pd.DataFrame = pd.read_csv('./data/checkpoints/df_carga_v1.csv',sep=	';')
df_clima : pd.DataFrame = pd.read_csv('./data/checkpoints/df_wheather_v1.csv',sep=	';')

df_clima.head()
```

```python
## Realizando o cálculo dos dados médios
df_clima['data-hora'] = df_clima['Data Medicao'] + ' ' + df_clima['Hora Medicao'].astype(str)

df_clima['data-hora'] = pd.to_datetime(df_clima['Data Medicao'] + df_clima['Hora Medicao'].astype(str), format='%Y-%m-%d%H')
df_clima.tail()

```

```python
# Renomeando as colunas
df_clima_renamed : pd.DataFrame = df_clima.copy()
df_clima_renamed.rename(
columns={
    'PRECIPITACAO TOTAL, HORARIO(mm)':'precipitacao_total',
    'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA(mB)':'pressao_atmosferica',
    'RADIACAO GLOBAL(Kj/m²)':'radiacao_global',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA(°C)':'temperatura_ar',
    'TEMPERATURA DO PONTO DE ORVALHO(°C)':'temperatura_ponto_orvalho',
    'UMIDADE RELATIVA DO AR, HORARIA(%)':'umidade_relativa_ar',
    'VENTO, DIRECAO HORARIA (gr)(° (gr))':'vento_direcao',
    'VENTO, VELOCIDADE HORARIA(m/s)':'velocidade_vento',
    'Data Medicao':'data_medicao',
    'Hora Medicao':'hora_medicao',
}, inplace=True)

df_clima_renamed : pd.DataFrame = df_clima_renamed[['data_medicao','hora_medicao','data-hora','precipitacao_total','pressao_atmosferica','radiacao_global','temperatura_ar','temperatura_ponto_orvalho','umidade_relativa_ar','vento_direcao','velocidade_vento']]
df_clima_renamed.head()
```

```python
def ajust_corrupted_float_value(value : str) -> float:
    """
    Função para corrigir valores com ponto flutuante que não estão no formato correto.

    Parameters
    ----------
    value : str
        Valor a ser corrigido.

    Returns
    -------
    float
        Valor corrigido.
    """
    try:
        value : str = str(value)
        if ',' in value:
            value : str = value.replace(',', '.')
        return float(value)
    except ValueError:
        if value[0] == ',':
            value : str = '0' + value
            value : float = float(value)
        else:
            raise ValueError("Erro ao converter precipitacao_total : Situação não prevista")
            
        return value
```

```python
## Ajustando os valores não inteiros de precipitacao_total
df_clima_renamed['precipitacao_total'] = df_clima_renamed.apply(lambda x: ajust_corrupted_float_value(x['precipitacao_total']), axis=1)
df_clima_renamed['velocidade_vento'] = df_clima_renamed.apply(lambda x: ajust_corrupted_float_value(x['velocidade_vento']), axis=1)
```

```python
df_clima_renamed['pressao_atmosferica'] = df_clima_renamed.apply(lambda x: ajust_corrupted_float_value(x['pressao_atmosferica']), axis=1)
df_clima_renamed['radiacao_global'] = df_clima_renamed.apply(lambda x: ajust_corrupted_float_value(x['radiacao_global']), axis=1)
df_clima_renamed['temperatura_ar'] = df_clima_renamed.apply(lambda x: ajust_corrupted_float_value(x['temperatura_ar']), axis=1)
```

```python
df_clima_renamed['temperatura_ponto_carvalho'] = df_clima_renamed.apply(lambda x: ajust_corrupted_float_value(x['temperatura_ponto_carvalho']), axis=1)
```

```python
df_clima_renamed.head()
df_clima_renamed.to_csv('./data/checkpoints/df_clima_v2.csv', sep=';', index=False)
```

```python
df_clima_renamed.head()
```

```python
## Sumarizando as informações

df_clima_grouped : pd.DataFrame = df_clima_renamed.groupby('data-hora').agg({
    'precipitacao_total':'mean',
    'pressao_atmosferica':'mean',
    'radiacao_global':'mean',
    'temperatura_ar':'mean',
    'temperatura_ponto_carvalho':'mean',
    'umidade_relativa_ar':'mean',
    'vento_direcao':'mean',
    'velocidade_vento':'mean',
})

df_clima_grouped.to_csv("./data/checkpoints/df_clima_grouped_mean.csv")
df_clima_grouped.head()
print(df_clima_grouped.columns)
```
